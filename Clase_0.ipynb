## Definición derivada

La herramienta básica que usaremos a lo largo del curso es la derivada. 
Esta se define como la tasa de cambio instantánea de una función. 

Sea  $f:\mathbb{R}\longrightarrow \mathbb{R}$. La tasa de cambio de $f$, ante una variación $\Delta x = h$ en $x$, se define como 

$\frac{\Delta y}{\Delta x} = \frac{f(x+h) - f(x)}{(x+h)-x} = \frac{f(x+h)-f(x)}{h}$

Esta tasa de cambio es la pendiente de la recta secante entre $\left(x,f(x)\right)$ y $\left(x+\Delta x,f(x+\Delta x)\right)$

En la medida en que $\Delta x \rightarrow 0$, esta recta secante se aproxima a la tangente en $\left(x,f(x) \right)$. De esta forma, la deriviada se define como



$\frac{d}{dx}\left( f(x) \right) = \lim_{h\rightarrow 0} \frac{f(x+h)-f(x)}{x+h-x}$

## Reglas de derivación y ejemplos

Algunas reglas de derivación:

* $\frac{d}{dx}\left( c \right) = 0$
* $\frac{d}{dx}\left( x^\alpha \right) = \alpha x^{\alpha-1}$
* $\frac{d}{dx}\left( \log(x) \right) = \frac{1}{x}$
* $\frac{d}{dx}\left( f(x) \cdot g(x)\right) = \frac{d}{dx}\left( f(x) \right) \cdot g(x)+ \frac{d}{dx}\left( g(x) \right) \cdot f(x)$
* $\frac{d}{dx}\left( f\left(g(x)\right) \right) = \frac{d}{dx}\left( f \right)\left(g(x)\right)\cdot \frac{d}{dx}\left( g(x) \right)$
* $\frac{d}{dx}\left( e^x \right) = e^x$

Ejemplos:

1. $\frac{d}{dx}\left( e^{x^{\alpha}} \right) = \alpha x ^{\alpha-1} e^{x^\alpha}$  
    
2. $\frac{d}{dx}\left( \log(x+x^2)\cdot e^{h(x)} \right)$

3. $\frac{d}{dx}\left( \frac{1}{\log\left(x^2 + xe^{x}\right)} \right)$

4. $\frac{d}{dx}\left( e^{\log(x^2)} \right) = \frac{d}{dx} x^2 = 2 x$


# Cálculo en varias variables

Veamos el caso ahora de una función de dos variables.

$\Phi: \mathbb{R} \times \mathbb{R} \rightarrow \mathbb{R}$ 

¿Se puede seguir hablando de los conceptos de derivada, tangente, función creciente/decreciente, cóncava/convexa?

Ahora usaremos el concepto de **derivada parcial**. Esta herramienta nos permite estudiar el crecimiento de la función $\Phi$ en una dirección particular.

$\frac{\partial\Phi(x,y)}{\partial x} = \lim_{h\rightarrow 0} \frac{\Phi(x+h,y)-\Phi(x,y)}{h}$

En otras palabras, *derivamos con respecto a x y dejamos todo lo demás constante*. Esta expresión se repetirá mucho a lo largo del curso.

Ejemplos

* $\frac{\partial }{\partial x} \left( x^{\alpha} y ^{1-\alpha} \right) = \alpha \left(\frac{y}{x}\right) ^ {1-\alpha}$

* $\frac{\partial}{ \partial x}\left( \sin\left(\log(xy)\right) \right) = y\frac{\cos\left(\log(xy)\right)}{x}$


* $\frac{\partial }{ \partial y}\left( x^2y-\frac{x}{y} \right) = x^2 +\frac{x}{y^2}$

* $\frac{\partial }{\partial y}\left( x^2 + 2\right) = 0$

¿Qué pasa si queremos saber la variación de $\Phi(x,y)$ ante un cambio en **ambas variables**? La respuesta es el llamado **diferencial total**

$d \Phi = \frac{\partial \Phi}{\partial x}dx +\frac{\partial \Phi}{\partial y}dy$

En este caso, ya no hablaremos de ''recta tangente'', sino que hablaremos de un ''plano (o hiperplano) tangente'', cuyos vector normal es $\left(  \frac{\partial \Phi}{\partial x} , \frac{\partial \Phi}{\partial y}\right)$

Por ejemplo, si $\Phi(x,y) = \log(x^2 +xy)$, entonces el diferencial total es

$d \Phi = \frac{1}{x^2 +xy}\left( 2x +y\right)dx +\frac{1}{x^2 +xy}\left( x\right)dy$



Por último, estudiaremos el concepto de **curva de nivel**.

$\frac{dy}{dx} = -\frac{\frac{\partial \Phi}{\partial x}}{\frac{\partial \Phi}{\partial y}}$

# Optimización

## Optimización sin restricciones

**Problema:** Dada una función $f:\mathbb{X}\rightarrow \mathbb{R}$, ¿existe un $x_0\in\mathbb{X}$ tal que $f(x_0)\geq f(x)    \forall x\in \mathbb{X}$? (Problema del máximo)

**Nota: En lo sucesivo supondremos, a menos que se explicite lo contrario, que estamos en un problema de maximización**

* Una condición necesaria para que $x_0$ sea un punto máximo es que $x_0$ sea un punto estacionario, es decir

$f'(x_0) = 0$ (Recordemos que si $\mathbb{X}\subseteq \mathbb{R}^N$, $f'$ corresponde al **gradiente** $\nabla f$).

Una demostración intuitiva de esta condición, es que si $f'(x)> 0$ significa que la función crece en la medida en que $x$ crece, por lo que $x$ no puede ser el máximo. Por otra parte, si $f'(x) <0$ significa que la función crece en la medida en que $x$ decrece, por lo que $x$ no puede ser el máximo. 
 

```python
def my_cubic(x):
    return (1/3)*x**3 + 1/2 * x**2 -2 * x

def der_cubic(x):
    return x**2 +x -2

#interact(graph_der, f1 = fixed(my_cubic), f2 = fixed(der_cubic), x1=(-4,4,0.01), lim = fixed(0), xmin = fixed(-5),
#        xmax = fixed(5), eje_x = fixed(r'x'), eje_y=fixed(r'f(x)'))

graph_der( f1 = my_cubic, f2 = der_cubic, x1=-3, lim = 0, xmin = -5,
        xmax = 5, eje_x = r'x', eje_y=r'f(x)')
```

Ahora, $f'(x_0) = 0$ es una condición necesaria, no suficiente. Esta condición es válido tanto para máximos como mínimos. Para garantizar que la función sea un máximo, necesitamos que la función sea estrictamente cóncava en ese punto.

$f''(x_0) < 0$

Como vemos en el gráfico, en el punto máximo la primera derivada es decreciente (o sea que la segunda derivada es negativa).

## Optimización con restricciones

Cuando estamos en presencia de restricciones, los conceptos anteriores ya no se pueden aplicar directamente.

Supongamos que nos enfrentamos al siguiente problema

$\max_{x} x^2 \mbox{  s.a.  } x\in [-0.5,1]$

Esta función tiene un punto estacionario en $x= 0$, pero claramente no es el máximo de $x^2$.

```python
def my_square(x):
    return x**2

fig, ax = plt.subplots(figsize=(5,5))
x = np.linspace(-1,1.5)
ax.plot(x,my_square(x))
ax.axvline(x = 1, ymin = 0, ymax = 1, color = 'C1')
ax.axvline(x = -0.5, ymin = 0, ymax = 1, color = 'C1')
ax.scatter(1,my_square(1), color = 'C3')
#plt.ylim(-0.1,2)
plt.show()
```

Formalmente, el problema consiste en

Sea $f:\mathbb{X}\subseteq \mathbb{R}^N\rightarrow \mathbb{R}$ la función a maximizar. 

Sea $g:\mathbb{X}\subseteq \mathbb{R}^N\rightarrow \mathbb{R}^M$ un vector de $M$ restricciones.

El problema de maximización es

$\max_x f(x) \mbox{ s.a. } g(x) \leq 0$


En el ejemplo anterior
en el que $x\in [-0.5,1]$, tenemos que $f(x) = x^2$ y $g(x) = \begin{pmatrix}-x+(-0.5) \\x-1 \end{pmatrix}$

Para resolver este problema, distinguiremos dos tipos de restricciones. Las que se cumplen con igualdad o con desigualdad.

1. Si $g(x)=0$, "despejamos" $M$ variables del sistema de $M$ ecuaciones $g(x)=0$ y las reemplazamos en la función $f$, luego derivamos e igualamos a cero.
   
   Por ejemplo:
   
   $\max_{x,y} \log(y) + x$ s.a. $g(x,y) = x + y - A= 0$
   
   En este caso $N=2$ y $M=1$. Despejamos 1 variable de la ecuación $g(x,y) = 0$ y la reemplazamos en $f(x,y) = \log(y) + x$.
   
   $x = A-y$
   
   $f(x,y) = f(A-y,y) = \log(y)+A-y = \tilde{f}(y)$
   
   Luego, el problema se reduce a $\max_{y} \log(y)+A-y$
   
   Para resolver derivamos e igualamos a cero.
   $\frac{\partial}{\partial y}\tilde{f}(y) = \frac{1}{y}-1 = 0$, de donde $y = 1$. ¿Qué pasa con $x$? Reemplazamos
   $y=1$ en $g(x,y) = 0$.
   
   $g(x,1)=x+ 1 -A = 0$, luego $x = A-1$.
   
   
2. Si $g(x) \leq 0$, usaremos la técnica de los "multiplicadores de Lagrange".
   
   Definimos la función 
   
   $\mathcal{L}(x,\lambda) = f(x) -\lambda \cdot g(x)$.
   
   Esta función se conoce como "lagrangeano". $\lambda$ es un vector con $M$ componentes (uno por cada restricción), es decir $\lambda =(\lambda_1,...,\lambda_M)$
   
   Ahora, la condición de optimalidad es
   
   * $\frac{\partial }{\partial x} \mathcal{L}(x) = \nabla f -\lambda \nabla g =  0$
   * $\lambda_m\cdot g_m(x) = 0$  $\forall m \leq M$
   * $\lambda_m \geq 0$  $\forall m \leq M$
   
   
   **Ejemplo!**
   
   Sea $f(x,y,z) = xyz$ y $g(x,y,z) = x+y-1$. Se pide resolver
   
   $\max_{\{x,y,z\}} f(x,y,z)$ s.a. $g(x,y,z)\leq 0$ 
   
   
   **Respuesta:**
   
   Definimos el lagrangeano y lo derivamos con respecto a $(x,y,z)$
   
   $\mathcal{L}(x,y,z,\lambda) = xyz -\lambda (x^3+y^3+z^3-1)$
   
   $\frac{\partial}{\partial x}\mathcal{L} = yz -3 \lambda x^2 = 0$
   
   $\frac{\partial}{\partial y}\mathcal{L} = xz -3 \lambda y^2 = 0$
   
   $\frac{\partial}{\partial z}\mathcal{L} = xy -3 \lambda z^2 = 0$
   
   
   Multiplicamos la primera ecuación por $x$, la segunda por $y$ y la tercera por $z$
   

Resolver:

$\max_{x,y} x\cdot y $ sujeto a $x + y = A$

* Despejo $x$

$x = A -y$

* Reemplazo $x$ en la función objetivo

$\max_y (A-y)\cdot y  = \max_y Ay -y^2$

* Ahora derivamos con respecto a $y$

$A -2 y = 0$

* Despejar $y$

$y^* = A/2$

* Despejar $x$

Reemplazo en la restricción original $g(x,y) = x+y -A = 0$

$x + y^* -A = 0$

$x+ A/2-A = 0$

$x -A/2 = 0$

$x^* = A/2$

```python
def graph_der(f1,f2,x1 = -3, xmin = -5, xmax = 5, lim = 0, eje_x=r'x', eje_y=r'f(x)'):
    """
    Grafica una función señalando la recta tangente en determinado punto
    f1 -> Función a graficar
    f2 -> Derivada de f1
    x1 -> Punto en el que se grafica la recta tangente
    xmin -> límite inferior en el eje x
    xmax -> límite superior en el eje y
    lim = 0 -> deja libre los límites del gráfico
        = 1 -> eje x e y tienen los mismos límites
    """
    
    # Valores de x a ser evaluados
    x = np.linspace(xmin,xmax,100)

    # Punto en el que se grafica la tangencia (x1,y1)
    y1 = f1(x1)

    # Ecuación de la recta tangente a f1 en x1
    def line(x, x1, y1):
        return f2(x1)*(x - x1) + y1

    # Rango en el que se graficará la recta tangente
    # La recta tangente se gráfica en un radio 2 c/r a x1
    xrange = np.linspace(x1-2, x1+2, 10)

    # Gráfico
    
    fig, ax = plt.subplots(figsize=(7,7))
    
    # Leyenda del gráfico
    labels =[]
    utext1 = r'$f({:5.1f}) = {:5.1f} $'.format(x1,f1(x1)) # f(x1) = y1
    labels.append(utext1)
    utext2 = r'$df/dx({:5.1f}) = {:5.1f}$'.format(x1,f2(x1)) #f1'(x1) = y1'
    labels.append(utext2)
    handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", 
                                 lw=0, alpha=0)] * 2
    ax.legend(handles, labels, loc='best', fontsize='small', 
          fancybox=True, framealpha=0.7, 
          handlelength=0, handletextpad=0)
    
    if lim == 1:
        ax.set_xlim(xmin, xmax)
        ax.set_ylim(xmin, xmax)
    plt.xlabel(eje_x)
    plt.ylabel(eje_y)
    ax.plot(x, f1(x)) # Función primitiva
    ax.scatter(x1, y1, color='C1', s=50) # Punto a evaluar 
    ax.plot(xrange, line(xrange, x1, y1), 'C1--', linewidth = 2) # Recta tangente
```

```python
def approx_tg(f1,f2,x1 = -3,h = 0.05):
    """
    Aproximación a la tangente
    f1 -> Función a graficar
    f2 -> Derivada de f1
    x1 -> Punto en el que se grafica la recta tangente
    h -> variación de x / nivel de aproximación
    """
    
    # Preimagen de f1
    x = np.linspace(-5,5,100)

    #Punto en el que se grafica la tangencia (x1,y1)
    y1 = f1(x1)

    # Ecuación de la recta tangente en x1
    def line(x, x1, y1):
        return f2(x1)*(x - x1) + y1
    # Rango en el que se graficará la recta tangente
    xrange = np.linspace(x1-2, x1+2, 10)
    

    # Ecuación de la recta secante en x1 y x1+h
    m = ((f1(x1+h)-f1(x1))/(x1+h-x1))
    def sec_line(x,x1):
        c = f1(x1) -m*x1
        return m*x + c
    xrange_sec = np.linspace(x1-2, np.min([x1+h+2, 5]), 100)
    
    # Gráfico
    fig, ax = plt.subplots(figsize=(7,7))
    
    # Leyenda del gráfico
    labels =[]
    utext1 = r'$df/dx({:5.1f}) = {:5.1f} $'.format(x1,f2(x1)) # Derivada real 
    labels.append(utext1)
    utext2 = r'$\Delta f/ \Delta x({:5.1f}) = {:5.1f}$'.format(x1,m) # Aproximación con la secante
    labels.append(utext2)
    utext3 = r'$\Delta x =  {:5.1f}$'.format(h) # Variación de x
    labels.append(utext3)
    handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", lw=0, alpha=0)] * 3
    ax.legend(handles, labels, loc='best', fontsize='small', fancybox=True, framealpha=0.7, handlelength=0, handletextpad=0)

    ax.plot(x, f1(x))
    ax.scatter(x1, y1, color='C1', s=50) # Punto en el que se calcula la derivada
    ax.scatter(x1+h,f1(x1+h),color='C2') # Punto por el que pasa la recta secante
    ax.plot(xrange, line(xrange, x1, y1), 'C1--', linewidth = 2) # Recta tangente
    ax.plot(xrange_sec, sec_line(xrange_sec,x1)) # Recta secante
```

```python
def multi_der_plots(f1,f2,f3, x1 = -3, xmin=-5, xmax=5, lim=0):
    """
    f1 = Función primitiva
    f2 = Primera derivada
    f3 = Segunda derivada
    """

    x = np.linspace(xmin,xmax,100)

    #Punto en el que se grafica la tangencia (x1,y1)
    y1 = f1(x1)
    y2 = f2(x1)

    # Definimos recta tangente en x1
    def line(x, x1, y1,f2):
        return f2(x1)*(x - x1) + y1

    # Rango en el que se graficará la recta tangente
    xrange = np.linspace(x1-2, x1+2, 10)


    fig, axs = plt.subplots(3, figsize=(7,11))

    handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", 
                                     lw=0, alpha=0)] * 4

    axs[0].plot(x, f1(x))
    axs[0].scatter(x1, y1, color='C1', s=50)
    axs[0].plot(xrange, line(xrange, x1, y1,f2), 'C1--', linewidth = 2)
    if lim == 1:
        axs[0].set_xlim(xmin, xmax)
        axs[0].set_ylim(xmin, xmax)

    utext1 = r'$f({:5.1f}) = {:5.1f} $'.format(x1,f1(x1))
    axs[0].text(xmin,np.max(f1(x))-0.05*(max(f1(x))-min(f1(x))),utext1,fontsize=10)
    axs[0].axvline(x=x1,ymin=-1.2,ymax=1,c="red",linewidth=2,zorder=0, clip_on=False)


    axs[1].plot(x, f2(x))
    axs[1].scatter(x1, y2, color='C1', s=50)
    axs[1].plot(xrange, line(xrange, x1, y2,f3), 'C1--', linewidth = 2)
    if lim == 1:
        axs[1].set_xlim(xmin, xmax)
        axs[1].set_ylim(xmin, xmax)
    utext2 = r'$df({:5.1f})/dx = {:5.1f} $'.format(x1,f2(x1))
    axs[1].text(xmin,np.max(f2(x))-0.05*(max(f2(x))-min(f2(x))),utext2,fontsize=10)
    axs[1].axvline(x=x1,ymin=-1.2,ymax=1,c="red",linewidth=2,zorder=0, clip_on=False)

    axs[2].plot(x, f3(x))
    axs[2].scatter(x1, f3(x1), color='C1', s=50)
    if lim == 1:
        axs[2].set_xlim(xmin, xmax)
        ax[2].set_ylim(xmin, xmax)
    utext3 = r'$d^2f({:5.1f})/dx^2 = {:5.1f} $'.format(x1,f3(x1))
    axs[2].text(xmin,np.max(f3(x))-0.05*(max(f3(x))-min(f3(x))),utext3,fontsize=10)
    axs[2].axvline(x=x1,ymin=-0.0,ymax=1,c="red",linewidth=2,zorder=1, clip_on=False)
```

```python
def plot_nivel(u_func):
    from matplotlib import cm
    
    N = 100 # number of elements
    x_max = 10 # maximum value
    x_vec = np.linspace(0,x_max,N)#[::-1]
    x1_values,x2_values = np.meshgrid(x_vec,x_vec,indexing='ij')
    
    u_values = u_func(x1_values,x2_values).T
    
    fig, ax = plt.subplots(figsize=(7,7))
    
    #plt.figure(figsize=(9,9))
    ax.matshow(u_values, cmap=cm.jet)

def plot_3d(u_func):

    N = 100 # number of elements
    x_max = 10 # maximum value

    x_vec = np.linspace(0,x_max,N)
    x1_values,x2_values = np.meshgrid(x_vec,x_vec,indexing='ij')
    u_values = u_func(x1_values,x2_values)

    # a. actual plot
    fig = plt.figure(figsize=(5,5))
    ax = fig.add_subplot(1,1,1,projection='3d')
    ax.plot_surface(x1_values,x2_values,u_values,cmap=cm.jet)

    # b. add labels
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_zlabel('$u$')

    # c. invert xaxis
    ax.invert_xaxis()

    # d. save
    fig.tight_layout()
    fig.savefig('someplot.pdf') # or e.g. .png 
```

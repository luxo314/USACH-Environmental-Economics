# Externalidades negativas

<p>Vimos que en presencia de externalidades no se cumple el primer teorema del bienestar. Esto significa que el mercado ya no nos lleva a una asignación eficiente en el sentido de Pareto. En consecuencia, el equilibrio de mercado podría no estar maximmizando el excedente total de la economía.</p>
<p>En este ejemplo, vemos que existe una externalidad negativa asociada a la producción de este bien. Siendo el costo marginal social la
    suma del costo marginal privado (la oferta) y la externalidad negativa marginal.</p>
<p>En el primer gráfico vemos la suma de ambos excedentes. Sin embargo, si consideramos el costo marginal externo, vemos que
el costo de la producción estaba claramente subestimado. De hecho, compensa buena parte del excedente 
del productor y del consumidor.</p>

```python
import EcoAmb as EA
def P_d(Q):
    return 0.5/(Q+0.1)
def P_s(Q):
    return Q**2

Mercado = EA.Market(P_d,P_s, Extm_P = 0.5)
```

```python
fig = plt.figure(figsize = (8,4))
plt.subplots_adjust(hspace = 0.6 )
fig.add_subplot(121)
Mercado.Plot_Equilibrio(mostrar_EC = True,mostrar_EP = True, titulo = r'Excedentes del productor y del consumidor')
fig.add_subplot(122)
Mercado.Plot_Equilibrio(mostrar_CMS = True, titulo = r'Costo social')
```

<p>Vemos que la asignación de mercado no es óptima ya que cada unidad producida por encima del óptimo social, genera un costo
mayor a la disposición a pagar por el bien.</p>
<p>De esta forma, la pérdida de bienestar es, simplemente, el área entre la curva de CMGS y la curva de demanda en el tramo $\left[Q^s,Q^m\right]$</p>

```python
Mercado.Plot_Equilibrio(mostrar_DWL = True, titulo = r'Pérdida de bienestar')
```

# Externalidad positiva

<p>Acá, por el contrario, vemos que el costo está subestimado. En efecto, en el equilibrio de mercado la última unidad
producida aporta un beneficio mayor que el costo que le genera a la sociedad.</p>
<p>De esta forma, la pérdida que sufre la economía es el área de la diferencia entre la curva de demanda y la curva de costo
    marginal social en el tramo $\left[Q^m,Q^s\right]$

```python
import EcoAmb as EA

Mercado2 = EA.Market(P_d,P_s, Extm_P = -0.25)

fig = plt.figure(figsize = (8,4))
plt.subplots_adjust(hspace = 0.6 )
fig.add_subplot(121)
Mercado2.Plot_Equilibrio(mostrar_EC = True,mostrar_EP = True, titulo = r'Excedentes del productor y del consumidor')
fig.add_subplot(122)
Mercado2.Plot_Equilibrio(mostrar_CMS = True, titulo = r'Costo social')
```

```python
Mercado2.Plot_Equilibrio(mostrar_DWL= True, titulo = r'Pérdida de bienestar')
```

<p>Ejemplo de externalidades positivas pueden ser:</p>
<ul>
    <li>Educación</li>
    <li>Construcción de obras públicas</li>
    <li>Vacunación</li>

Tanto las externalidades positivas como negativas, no solo aplican a los mercados "tradicionales" con oferta y demanda, sino que aplican 
a culaquier nivel de intercambio.

| jugador1/Jugador2        | Delatar        | Callar       |
| ------------- |:-------------: | -----:       |
| **Deltar**        |(4 años, 4 años)| (0, 10 años) |
| **Callar**        | (10 años, 0)   |(1 mes, 1 mes)|

<p>Este es el llamado "dilema del prisionero". Notemos primero que el "excedente" o bienestar total da ambos prisioneros se maximiza si ambos prisioneros cooperan entre sí y deciden callar, de esta forma, cada uno solo recibe un mes de carcel.</p>
<p>Sin embargo, la maximización individual <strong>no nos lleva</strong> al resultado anterior. Veamos por qué:</p>

<ul>
    <li>Supongamos que ambos prisioneros deciden callar y lograr que cada uno tenga un mes de cárcel</li>
    <li>El prisionero 1 observa que si delata a su compañero, dado que este decidió callar, podrá salir libre de inmediato.</li>
    <li>En consecuencia, el prisionero 1 delatará a su compañero. El prisionero 2 realiza el mismo razonamiento</li>
    <li>En conclusión, la cooperación no es un equilibrio</li>
</ul>
<p>De hecho, el único equilibrio que hay en este caso es la no cooperación, obteniendo cada uno 4 años de cárcel.</p>
